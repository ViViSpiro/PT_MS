```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
y = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
```

```python
b = (np.mean(x * y) - np.mean(x) * np.mean(y)) / (np.mean(x**2) - np.mean(x)**2)
b
```

```python
a = np.mean(y) - b * np.mean(x)
a
```

```python
y_pred = a + b * x
y_pred
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred, c='r', label=r'$y=444,18+2,62\cdot x$')
plt.legend()
plt.xlabel('Величина заработной платы')
plt.ylabel('Поведенческий кредитный скоринг')
plt.show()
```

#### Без использования intercept:

```python
x_ = x.reshape((-1, 1))
x_
```

```python
y_ = y.reshape((-1, 1))
y_
```

```python
b_ = np.dot(np.linalg.inv(np.dot(x_.T, x_)), np.dot(x_.T, y_))
b_
```

```python
y_pred_ = b_ * x_
y_pred_
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred, c='r', label=r'$y=444,18+2,62\cdot x$')
plt.plot(x_, y_pred_, c='g', label=r'$y=5,89\cdot x$')
plt.legend()
plt.xlabel('Величина заработной платы')
plt.ylabel('Поведенческий кредитный скоринг')
plt.show()
```
